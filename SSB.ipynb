```python
#Load in table of Vehicles
url = "https://data.ssb.no/api/v0/no/table/07849/"
#Query
json_q = {
  "query": [
    {
      "code": "Region",
      "selection": {
        "filter": "vs:Landet",
        "values": []
      }
    },
    {
      "code": "DrivstoffType",
      "selection": {
        "filter": "item",
        "values": [
          "1",
          "2",
          "3",
          "4",
          "5",
          "6"
        ]
      }
    },
    {
      "code": "ContentsCode",
      "selection": {
        "filter": "item",
        "values": [
          "Personbil1"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}
       
```

```python
res = requests.post(url, json=json_q)
ds = pyjstat.Dataset.read(res.text) #Create df of results
df = ds.write('dataframe')
```

```python
df['drivstofftype'].unique()
```

```python
df_fuel = df[df['drivstofftype']=='Bensin'].groupby('år').sum()
df_diesel = df[df['drivstofftype']=='Diesel'].groupby('år').sum()
df_el = df[df['drivstofftype']=="El."].groupby('år').sum()
df_no_el = df[df['drivstofftype']!='El.'].groupby('år').sum()
```

```python
df_el = df[df['drivstofftype']=="El."].groupby('år').sum()
xvals = [int(i) for i in df_el.index]
yvals = [i for i in df_no_el['value']]
yvals_el = [i for i in df_el['value']]
```

```python
#Plot


plt.figure(figsize = (15,10))

fig1 = plt.plot(xvals, yvals_el, label="El Vehicles", )
plt.xlabel("Year", fontsize=20)
plt.ylabel("Number of Electric Vehicles", fontsize=20)
plt.margins(x=0)
plt.tick_params(axis='both', which='major', labelsize=20)

plt.savefig("Data/Figures/ElVehicles.png", dpi=400,bbox_inches='tight')
```

```python
plt.figure(figsize=(15,10), dpi=200)
fig2 =plt.plot(xvals, yvals, label="ICEV" ,color='orange')
fig1 = plt.plot(xvals, yvals_el, label="EV")

plt.fill_between(xvals,yvals, yvals_el, color="orange", alpha=0.3)
plt.fill_between(xvals, yvals_el, alpha=0.5)
plt.ticklabel_format(style='plain')
plt.xlabel("Year")
plt.ylabel("Number of Vehicles")
plt.margins(x=0)
plt.legend(bbox_to_anchor=(1.2,0.5), frameon=False)
plt.rcParams.update({'font.size': 25})
plt.grid()
plt.savefig("Data/Figures/EVvsICEV.jpg", dpi=400, bbox_inches="tight")
```
